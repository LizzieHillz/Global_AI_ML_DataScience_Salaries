```python
#1. What is the average salary for AI, ML, and data science professionals worldwide?
avg_salary = df.groupby(['job_group', 'experience_level'])['salary_in_usd'].mean().reset_index()
avg_salary_pivot = avg_salary.pivot(index= 'experience_level', columns= 'job_group', values= 'salary_in_usd')
print("Average Salary by Job Groups and Experience Level:")
avg_salary_pivot
```

```python
#Plotting a line graph to show average salaries of job groups across experience levels
avg_salary_pivot.plot(kind='line',marker='*',figsize =(10,6))
plt.xlabel('job groups')
plt.ylabel('Average Salaries')
plt.grid(True)
plt.title('Analysis of average salaries of different job groups across experience levels')
plt.legend(title='Experience levels')
plt.show()
```

```python
# 2. Overall Distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], kde=True)
plt.title('Overall Distribution of Salaries')
plt.xlabel('salary_in_usd')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Average, median, and mode of salaries
avg_salary = df['salary_in_usd'].mean()
median_salary = df['salary_in_usd'].median()
mode_salary = df['salary_in_usd'].mode()[0]
print(f"Average Salary: ${avg_salary:.2f}")
print(f"Median Salary: ${median_salary:.2f}")
print(f"Mode Salary: ${mode_salary:.2f}")
```

```python
# 4. Range of salaries
min_salary = df['salary_in_usd'].min()
max_salary = df['salary_in_usd'].max()
print(f"Salary Range: ${min_salary:.2f} - ${max_salary:.2f}")
```

```python
# Geographic and Industry Analysis

# 1a. Regions offering the highest salaries and contributing factors
region_avg_salary = df.groupby(['Region', 'job_group'])['salary_in_usd'].mean().reset_index().sort_values(by='salary_in_usd', ascending=True)
region_avg_salary_pivot = region_avg_salary.pivot(index= 'Region', columns= 'job_group', values= 'salary_in_usd')
print("Average Salary by Company's Region")
region_avg_salary_pivot
```

```python
# Creating a Bar chart for average salary by region and job group
plt.figure(figsize=(14, 8))
sns.barplot(x='Region', y='salary_in_usd', hue='job_group', data= region_avg_salary)
plt.title("Average Salary by Company's Region and Job Group")
plt.xlabel('Region')
plt.ylabel('Average Salary')
plt.legend(title='Job Group')
plt.show()
```

```python
# Correlation with salary
skills_salary_correlation = df.explode('job_group').groupby('job_group')['salary_in_usd'].mean().reset_index().sort_values(by='salary_in_usd', ascending=False)
print("Job Title with Highest Average Salaries:")
print(skills_salary_correlation)
```

```python
# Salary evolution over the years
salary_over_years = df.groupby('work_year')['salary_in_usd'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='work_year', y='salary_in_usd', data=salary_over_years)
plt.title('Salary Evolution Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.show()
```

```python
# Calculate the percentage of each employment type by year
employment_type_counts = df.groupby(['work_year', 'employment_type']).size().reset_index(name='count')
total_count_per_year = df.groupby('work_year').size().reset_index(name='total_count')
employment_type_percentage = employment_type_counts.merge(total_count_per_year, on='work_year')
employment_type_percentage['percentage'] = (employment_type_percentage['count'] / employment_type_percentage['total_count']) * 100

# Visualize employment type trends over the years
plt.figure(figsize=(12, 6))
sns.barplot(data=employment_type_percentage, x='work_year', y='percentage', hue='employment_type')
plt.title('Employment Type Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend(title='Employment Type')
plt.show()

# Calculate the percentage of remote work by year
remote_work_percentage_by_year = df.groupby(['work_year', 'remote_ratio']).size().reset_index(name='count')
remote_work_percentage_by_year = remote_work_percentage_by_year.merge(total_count_per_year, on='work_year')
remote_work_percentage_by_year['percentage'] = (remote_work_percentage_by_year['count'] / remote_work_percentage_by_year['total_count']) * 100

# Map remote_ratio to categorical labels for better visualization
remote_work_levels = {0: 'In-Office', 50: 'Hybrid', 100: 'Fully Remote'}
remote_work_percentage_by_year['remote_work_level'] = remote_work_percentage_by_year['remote_ratio'].map(remote_work_levels)

# Visualize remote work trends over the years
plt.figure(figsize=(12, 6))
sns.barplot(data=remote_work_percentage_by_year, x='work_year', y='percentage', hue='remote_work_level')
plt.title('Remote Work Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage of Professionals Working Remotely')
plt.legend(title='Remote Work Level')
plt.show()
```

```python
# Correlation and Advanced Analysis

# 1. Correlation between experience level and salaries
experience_salary_corr = df['experience_level'].astype('category').cat.codes.corr(df['salary_in_usd'])
print(f"Correlation between Experience Level and Salary: {experience_salary_corr:.2f}")
```

```python
#what is the relationship between experience level and salaries
experience = df['experience_level']
salary_usd = df['salary_in_usd']

plt.scatter(experience, salary_usd, alpha = 0.5)
plt.xlabel('experience_level')
plt.ylabel('salary_in_usd')
plt.show()
```

```python
# 2. Correlation between company size and salaries
company_size_salary_corr = df['company_size'].astype('category').cat.codes.corr(df['salary_in_usd'])
print(f"Correlation between Company Size and Salary: {company_size_salary_corr:.2f}")
```

```python
#what is the relationship between company size and salaries
company = df['company_size']
salary_usd = df['salary_in_usd']

plt.scatter(company, salary_usd, alpha = 0.5)
plt.xlabel('company_size')
plt.ylabel('salary_in_usd')
plt.show()
```

```python
# 3. Correlation between region and salaries
region_salary_corr = df['Region'].astype('category').cat.codes.corr(df['salary_in_usd'])
print(f"Correlation between Company_location and Salary: {region_salary_corr:.2f}")
```

```python
#what is the relationship between region and salaries
location = df['Region']
salary_usd = df['salary_in_usd']

plt.scatter(location, salary_usd, alpha = 0.5)
plt.xlabel('company_size')
plt.ylabel('salary_in_usd')
plt.show()
```

```python
# 4. Interaction of factors influencing salary
# How do different factors (e.g., job group, experience level, company size and location) interact to influence salary?
interaction_factors = df[['salary_in_usd', 'experience_level', 'company_size', 'Region', 'job_group']]

# Convert categorical variables to numerical
interaction_factors['experience_level'] = interaction_factors['experience_level'].astype('category').cat.codes
interaction_factors['company_size'] = interaction_factors['company_size'].astype('category').cat.codes
interaction_factors['Region'] = interaction_factors['Region'].astype('category').cat.codes
interaction_factors['job_group'] = interaction_factors['job_group'].astype('category').cat.codes

plt.figure(figsize=(10, 6))
sns.heatmap(interaction_factors.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Factors Influencing Salary')
plt.show()
```
